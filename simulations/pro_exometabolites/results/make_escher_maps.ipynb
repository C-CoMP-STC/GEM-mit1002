```python
import escher
import pandas as pd
import cobra
```

```python
# Load the model
model = cobra.io.read_sbml_model("../../../model.xml")
```

```python
# Save the path to the Escher map (with glycolysis and TCA cycle)
amac_map_path = "../../../escher/MIT1002_glycolysis_and_tca_escher-map.json"
```

```python
# Load the results
results = pd.read_csv("fba_results.csv")
```

```python
results
```

```python
# Make map with the solution for a single time point
builder = escher.Builder(
    model=model, map_json=amac_map_path
)
builder.reaction_data = eval(results["fluxes"][results["time_point"] == 0].iloc[0])
builder
# Then save the map manually from the widget
```
