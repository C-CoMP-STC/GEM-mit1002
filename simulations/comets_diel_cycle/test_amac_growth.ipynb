```python
import cobra
```

```python
model = cobra.io.read_sbml_model("../../model.xml")
```

```python
model.metabolites.cpd00229_c0
```

```python
# Make an external version of the glycoladehyde metabolite
gcald_e0_met = cobra.Metabolite("cpd00229_e0",
    compartment="e0",
    name=model.metabolites.cpd00229_c0.name,
    formula=model.metabolites.cpd00229_c0.formula,
    charge=model.metabolites.cpd00229_c0.charge)
gcald_e0_met
```

```python
# Add a transport reaction for glycoaldehyde
gcald_trans = cobra.Reaction("Gcald_transport")
gcald_trans.name = "Glycoaldehyde transport"
gcald_trans.lower_bound = -1000  # Allow uptake
gcald_trans.upper_bound = 1000   # Allow secretion
gcald_trans.add_metabolites({
    model.metabolites.cpd00229_c0: -1,  # Glycoaldehyde in the cytosol
    gcald_e0_met: 1,   # Glycoaldehyde in the extracellular space
})
model.add_reactions([gcald_trans])
model.reactions.Gcald_transport
```

```python
# Add an exchange reaction for the external glycoaldehyde
model.add_boundary(model.metabolites.cpd00229_e0, type="exchange")
model.reactions.EX_cpd00229_e0
```

```python
# Define a medium
medium = {
    # Glycoaldehyde is one of the main carbon sources produced by PRO
    "EX_cpd00229_e0": 10,  # Glycoaldehyde
    # Everything in the COMETS medium with ModelSEED IDs
    "EX_cpd00013_e0": 1000,  # Ammonia
    "EX_cpd00011_e0": 1000,  # CO2
    "EX_cpd00067_e0": 1000,  # H+
    "EX_cpd00009_e0": 1000,  # Phosphate
    "EX_cpd00001_e0": 1000,  # H2O
    "EX_cpd00063_e0": 1000,  # Ca2+
    "EX_cpd00099_e0": 1000,  # Cl-
    "EX_cpd00149_e0": 1000,  # Co2+
    "EX_cpd00058_e0": 1000,  # Cu2+
    # "EX_cpd10515_e0": 1000,  # Fe+2 (NOT IN Amac model)
    "EX_cpd00254_e0": 1000,  # Mg2+
    "EX_cpd00205_e0": 1000,  # K+
    "EX_cpd00971_e0": 1000,  # Na+
    # "EX_cpd09695_e0": 1000,  # Strontium (NOT IN Amac model)
    "EX_cpd00048_e0": 1000,  # Sulfate
    "EX_cpd00034_e0": 1000,  # Zn2+
    # "EX_cpd00239_e0": 1000,  # HS- / bisulfide  (NOT IN Amac model)
    # HAD TO ADD
    # TODO: Add these in the COMETS sim
    "EX_cpd10516_e0": 1000,  # Fe+3
    "EX_cpd00007_e0": 1000,  # Oxygen
    "EX_cpd00030_e0": 1000,  # Mn2+_e0
}
model.medium = medium
```
