```python
import cobra
```

```python
model = cobra.io.read_sbml_model("model.xml")
```

```python
cobra.manipulation.remove_genes(model, ['WP_012518362.1', 'WP_014949447.1', 'WP_014949446.1', 'WP_014949445.1'])
```

```python
# Remove the two step version of SUCD
model.remove_reactions([model.reactions.rxn00288_c0, model.reactions.rxn10126_c0])
```

```python
# Define the lumped SUCD reaction
new_rxn = cobra.Reaction("rxn09272_c0")
new_rxn.name = "Succinate dehydrogenase (ubiquinone)"
new_rxn.add_metabolites({
    model.metabolites.cpd00036_c0: -1,
    model.metabolites.cpd15560_c0: -1,
    model.metabolites.cpd00106_c0: 1,
    model.metabolites.cpd15561_c0: 1
})
```

```python
model.add_reactions([new_rxn])
```

```python
cobra.io.write_sbml_model(model, "model.xml")
```
