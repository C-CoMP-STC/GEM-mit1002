```python
import json

import cobra
import escher
```

```python
# Load the model
model = cobra.io.read_sbml_model("../model.xml")
```

```python
# Save paths to maps
# Paths relative to this notebook
cent_met_map = "../escher/MIT1002_glycolysis_and_tca_escher-map.json"
proton_map = "../escher/MIT1002_all-extracellular-proton-rxns_escher-map.json"
```

```python
# Load the flux data for growth on glucose and nitrate
# This is pFBA, generated in nitrate_vs_ammonia/check_nitrate_growth.ipynb
with open("fluxes_Glucose_Nitrate_p0.json") as f:
    flux_data = json.load(f)
```

```python
# Make map with the pFBA solution
builder = escher.Builder(
    model=model,
    map_json=cent_met_map
)
builder.reaction_data = flux_data
builder
# Then save the map manually from the widget
```

```python
# Make map with the pFBA solution
builder = escher.Builder(
    model=model,
    map_json=proton_map
)
builder.reaction_data = flux_data
builder
# Then save the map manually from the widget
```
