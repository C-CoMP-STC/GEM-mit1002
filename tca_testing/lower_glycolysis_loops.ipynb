```python
import escher
import cobra
```

```python
# Load the model
amac_model = cobra.io.read_sbml_model("../model.xml")
```

```python
# Save the paths to the maps
amac_map_path = "../escher/MIT1002_glycolysis_and_tca_escher-map.json"
```

```python
glc_nh4_media = {
    "EX_cpd00027_e0": 10,  # Glucose
    "EX_cpd00013_e0": 1000,  # NH4_e0
    "EX_cpd00058_e0": 1000,  # Cu2+_e0
    "EX_cpd00007_e0": 20,  # O2_e0
    "EX_cpd00971_e0": 1000,  # Na+_e0
    "EX_cpd00063_e0": 1000,  # Ca2+_e0
    "EX_cpd00048_e0": 1000,  # Sulfate_e0
    "EX_cpd10516_e0": 1000,  # fe3_e0
    "EX_cpd00254_e0": 1000,  # Mg_e0
    "EX_cpd00009_e0": 1000,  # Phosphate_e0
    "EX_cpd00205_e0": 1000,  # K+_e0
    "EX_cpd00099_e0": 1000,  # Cl-_e0
    "EX_cpd00030_e0": 1000,  # Mn2+_e0
    "EX_cpd00001_e0": 1000,  # H2O_e0
    "EX_cpd00034_e0": 1000,  # Zn2+_e0
    "EX_cpd00149_e0": 1000,  # Co2+_e0
}
```

```python
# Run normal FBA with glucose and NH4+ media
amac_model.medium = glc_nh4_media
fba_solution = amac_model.optimize()
```

```python
# Run pFBA with glucose and NH4+ media
pfba_solution = cobra.flux_analysis.pfba(amac_model)
```

```python
# Make map with the FBA solution
builder = escher.Builder(model=amac_model, map_json=amac_map_path)
builder.reaction_data = fba_solution.fluxes
```

```python
# Make map with the pFBA solution
builder = escher.Builder(
    model=amac_model, map_json=amac_map_path
)
builder.reaction_data = pfba_solution.fluxes
```
