```python
import cobra
import pickle as pkl
```

```python
from gem_utilities.media import clean_media
```

```python
model = cobra.io.read_sbml_model('model.xml')
```

```python
# Load the media definitions
with open("test/test_files/media/media_definitions.pkl", "rb") as f:
    media_definitions = pkl.load(f)
```

```python
mbm_glucose = media_definitions['mbm'].copy()
mbm_glucose['EX_cpd00027_e0'] = 10
```

```python
model.medium = clean_media(model, mbm_glucose)
sol_mbm_glucose = model.optimize()
sol_mbm_glucose
```

```python
# Get the flux of ammonia uptake
sol_mbm_glucose.fluxes['EX_cpd00013_e0']
```

```python
# Try growing on L1 glucose
l1_glucose = media_definitions['l1'].copy()
l1_glucose['EX_cpd00027_e0'] = 10
```

```python
model.medium = clean_media(model, l1_glucose)
sol_l1_glucose = model.optimize()
sol_l1_glucose
```

```python
# Add ammonia to the L1 glucose media
l1_glucose['EX_cpd00013_e0'] = 10
model.medium = clean_media(model, l1_glucose)
sol_l1_glucose_w_ammonia = model.optimize()
sol_l1_glucose_w_ammonia
```
