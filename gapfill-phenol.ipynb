```python
import cobra
```

```python
model = cobra.io.read_sbml_model("model.xml")
```

```python
model.metabolites.cpd00027_c0.annotation
```

```python
model.reactions.rxn00007_c0.annotation
```

```python
# Add step 1
# Need to just add the product
met_2hm = cobra.Metabolite(
    "cpd01644_c0",
    name = "2-Hydroxymuconate",
    formula = "C6H4O5",
    charge = -2,
    compartment="c0"
)
met_2hm.annotation = {"bigg.metabolite": "2hmc",
                      "kegg.compound": "C02501",
                      "metacyc": "CPD-8720",
                      "metanetx.chemical": "MNXM162276",
                      'inchikey': "QTHJXLFFFTVYJC-UHFFFAOYSA-L"}
# Define the reaction
step_1 = cobra.Reaction(
    "rxn01993_c0",
    name = "2-Hydroxymuconate semialdehyde:NAD+ 6-oxidoreductase",
    lower_bound=-1000)
step_1.annotation = {
    'ec-code': '1.2.1.32',
    'bigg.reaction': 'HMSD',
    'kegg.reaction': 'R02762',
    'metacyc': 'RXN-8527',
    'rhea': ['34219', '34220', '34221', '34222'],
    'metanetx.reaction': 'MNXR126712'
}
step_1.add_metabolites({
    model.metabolites.cpd00001_c0: -1,
    model.metabolites.cpd00003_c0: -1,
    model.metabolites.cpd00518_c0: -1,
    model.metabolites.cpd00004_c0: 1,
    model.metabolites.cpd00067_c0: 2,
    met_2hm: 1
})
step_1
```

```python
# Step 2
# Define the product met
met_4oc = cobra.Metabolite(
    "cpd02183_c0",
    name = "4-Oxalocrotonate",
    formula = "C6H4O5",
    charge = -2,
    compartment= "c0"
)
met_4oc.annotation = {"bigg.metabolite": "oxalc",
                      "kegg.compound": "C03453",
                      "metacyc": "CPD-13337",
                      "metanetx.chemical": "MNXM6068",
                      'inchikey': "QTHJXLFFFTVYJC-UHFFFAOYSA-L"}
# Define the reaction
step_2 = cobra.Reaction(
    "rxn02788_c0",
    name = "R03966",
    lower_bound=-1000)
step_2.annotation = {
    'ec-code': '5.3.2.6',
    'bigg.reaction': '4OT',
    'kegg.reaction': 'R03966',
    'metanetx.reaction': 'MNXR126540'
}
step_2.add_metabolites({
    met_2hm: -1,
    met_4oc: 1
})
step_2
```

```python
# Step 3
# Product is already in the model
# Define the reaction
step_3 = cobra.Reaction(
    "rxn01894_c0",
    name="4-Oxalocrotonate carboxy-lyase"
)
# Purposefully didn't put the lower bound since modelSEED says that is not reversible
step_3.annotation = {
    'ec-code': '4.1.1.77',
    'bigg.reaction': '4OD',
    'kegg.reaction': 'R02602',
    'metanetx.reaction': 'MNXR95044'
}
step_3.add_metabolites({
    model.metabolites.cpd00067_c0: -1,
    met_4oc: -1,
    model.metabolites.cpd00011_c0: 1,
    model.metabolites.cpd00461_c0: 1
})
step_3
```

```python
# Add all of the reactions
model.add_reactions([step_1, step_2, step_3])
```

```python
# Save the model
cobra.io.write_sbml_model(model, "model.xml")
```
