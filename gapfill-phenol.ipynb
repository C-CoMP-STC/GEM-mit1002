```python
import cobra
```

```python
model = cobra.io.read_sbml_model("model.xml")
```

```python
# Add an external metabolite for phenol
phenol_ext = cobra.Metabolite(
    id="cpd00127_e0",
    name=model.metabolites.cpd00127_c0.name + " (external)",
    formula=model.metabolites.cpd00127_c0.formula,
    charge=model.metabolites.cpd00127_c0.charge,
    compartment="e0")
model.add_metabolites(phenol_ext)
```

```python
# Add a transport reaction for phenol
# Not ATP dependent, as phenol can diffuse across the membrane
trans_rxn = cobra.Reaction("rxn39188_c0")
trans_rxn.name = "Phenol transport"
trans_rxn.add_metabolites({
    model.metabolites.cpd00127_c0: -1,
    model.metabolites.cpd00127_e0: 1})
trans_rxn.lower_bound = -1000
model.add_reactions([trans_rxn])
```

```python
# Add exchange reaction for phenol
model.add_boundary(model.metabolites.cpd00127_e0, type="exchange", lb=0)
```

```python
# Save the model
cobra.io.write_sbml_model(model, "model.xml")
```
